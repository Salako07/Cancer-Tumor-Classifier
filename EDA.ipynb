```python
import os
from PIL import Image

def load_images_from_folder(folder):
    images = []
    
    for filename in os.listdir(folder):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):  # Add more formats if needed
            img_path = os.path.join(folder, filename)
            img = Image.open(img_path)
            images.append(img)
    
    return images

# Now, loaded_images contains all the images from the specified folder
```

```python
#Load Benign classImages
folder_path = 'Standardized_Images\Benign'  
BI= load_images_from_folder(folder_path)
```

```python
#Load Malignant classImages
folder_path = 'Standardized_Images\Malignant'  
MI= load_images_from_folder(folder_path)
```

```python
#Load Normal class Images
folder_path = r'Standardized_Images\Normal'  
NI= load_images_from_folder(folder_path)
```

```python
#Function to view first five images
def show_first_five_images(images):
    if len(images) >= 5:
        # Create a figure with subplots
        fig, axes = plt.subplots(1, 5, figsize=(15, 5))  # 1 row and 5 columns

        for i in range(5):
            axes[i].imshow(images[i])
            axes[i].axis('off')  # Turn off axis numbers and ticks

        plt.show()
    else:
        print("Less than 5 images in the list.")
```

```python
#Visualizing the Standardized Benign Images
show_first_five_images(BI)
```

```python
#Visualizing the Standardized Malignant Class Images
show_first_five_images(MI)
```

```python
#Visualizing the Standardized Normal Class Images
show_first_five_images(NI)
```
